```python
#error is too low,it could be a case of overfitting so let's check that through cross validation
```

```python
scores 
```

```python
def display_scores(scores):
    print("scores:",scores)
    print("mean:",scores.mean())
    print("variance:",scores.std())
```

```python
display_scores(tree_rmse_scores)
```

```python
#lets do the same for linear model
lin_scores = cross_val_score(l,data_prepd,label,scoring = "neg_mean_squared_error",cv = 10)
lin_rmse_scores = np.sqrt(-lin_scores)
display_scores(lin_rmse_scores)
```

```python
#we can see that this model was overfitting
#lets try another model
```

```python
from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor()

forest_reg.fit(data_prepd,label)
data_predictions = forest_reg.predict(data_prepd)
forest_mse = mean_squared_error(label,data_predictions)
forest_rmse = np.sqrt(forest_mse)
forest_rmse
```

```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline 
plt.style.use("ggplot")
```

```python
data = pd.read_csv("data.csv")
data_num= data[[x for x in data.columns if data[x].dtypes in ["float64","int64"]]]
data_num
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(data_num)
list(zip(data_num.columns,pca.explained_variance_ratio_))
```

```python
from sklearn.decomposition import PCA
import numpy as np 
pca = PCA()
pca.fit(data_num)
cumsum = np.cumsum(pca.explained_variance_ratio_)
cumsum
```

```python
#we can see that first four attributed explain all the variance
#data_num2 = data_num.drop(["year","popularity","tempo","speechiness"],axis =1)
data_num2= data_num[["valence","year","acousticness","danceability"]]
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import silhouette_score
ss = MinMaxScaler()
data_scaled =ss.fit_transform(data_num2)
data_scaled
```

```python
data_num1 = pd.DataFrame(data = data_scaled ,index = range(1,len(data_scaled)+1),columns = data_num2.columns)
data_num1
```

```python
#lets decide the no. of k
s_score = []
s_score1 = []
for i in range(2,30):
    kmeans = KMeans(n_clusters = i)
    y_pred = kmeans.fit_predict(data_num1.iloc[:2000])
    s1 = silhouette_score(data_num1.iloc[:2000],kmeans.labels_)
    s = kmeans.inertia_
    s_score.append(s)    
    s_score1.append(s1)
fig,ax = plt.subplots(1,2,figsize = (20,5))
plt.subplot(121)
plt.plot(range(2,30),s_score)
plt.xticks(ticks =range(2,30),labels = range(2,30) )
plt.xlabel("no.of k")
plt.ylabel("inertia")
plt.subplot(122)
plt.plot(range(2,30),s_score1)
plt.xticks(ticks =range(2,30),labels = range(2,30) )
plt.xlabel("no.of k")
plt.ylabel("silhouette_score")
```

```python
#if we choose k = 20
kmeans = KMeans(n_clusters = 20)
y_pred = kmeans.fit_predict(data_num1)
data_num1["predicted_genre"]= y_pred
data_num1
```

```python
#now lets see no. of songs in each predicted genre
genre_size = data_num1.groupby("predicted_genre").size()
genre_size.plot(kind = "bar",sort_columns = True)
```
